```python
models = ["Hermes 70b Q5", "Llama 3.3 70b", "Mistral 7b Q6"]
without_rag = [0.52, 0.52, 0.45]
with_rag = [0.54, 0.61, 0.53]

# Bar width and positions
x = np.arange(len(models))
width = 0.3

# Create figure and axis
fig, ax = plt.subplots(figsize=(11, 7))

# Plot bars
bars1 = ax.bar(x - width/2, without_rag, width,
               label='Without RAG', color='cyan')
bars2 = ax.bar(x + width/2, with_rag, width, label='RAG', color='blue')

# Labels and title
ax.set_xlabel("Models")
ax.set_ylabel("Accuracy")
ax.set_title("Performance Comparison: With and Without RAG")
ax.set_xticks(x)
ax.set_xticklabels(models, rotation=15)
ax.legend()

# Show values on top of bars
for bars in [bars1, bars2]:
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height:.2f}',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 6),  # Increased offset for better visibility
                    textcoords="offset points",
                    fontsize=10,
                    fontweight='bold',
                    ha='center', va='bottom',
                    color='black')

# Show plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Classification Report Data
categories = ["hatespeech", "normal", "offensive"]
precision = [0.63, 0.75, 0.51]
recall = [0.87, 0.40, 0.57]
f1_score = [0.73, 0.52, 0.54]
support = [305, 316, 309]

# Bar Chart for Precision, Recall, and F1-score
x = np.arange(len(categories))
width = 0.25

fig, ax = plt.subplots(figsize=(8, 5))
ax.bar(x - width, precision, width, label='Precision', color='cyan')
ax.bar(x, recall, width, label='Recall', color='blue')
ax.bar(x + width, f1_score, width, label='F1-score', color='Darkblue')

ax.set_xlabel("Categories")
ax.set_ylabel("Scores")
ax.set_title("Classification Report Metrics")
ax.set_xticks(x)
ax.set_xticklabels(categories)
ax.legend()

# Show values on top of bars
for bars in ax.containers:
    ax.bar_label(bars, fmt='%.2f', padding=3)

plt.tight_layout()
plt.show()

# Heatmap for Classification Report
fig, ax = plt.subplots(figsize=(8, 5))
data = pd.DataFrame({"Precision": precision, "Recall": recall,
                    "F1-score": f1_score}, index=categories)
sns.heatmap(data, annot=True, cmap="Blues", fmt=".2f", linewidths=0.5, ax=ax)
ax.set_title("RAG_llama3.3:70b-instruct-q6_K Classification Report")

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Training Data Distribution
train_labels = ["normal", "hatespeech", "offensive"]
train_counts = [4287, 3409, 1304]
test_labels = ["hatespeech", "normal", "offensive"]
test_counts = [628, 628, 628]

# Plot Train and Test Distribution
fig, ax = plt.subplots(1, 2, figsize=(12, 5))
ax[0].bar(train_labels, train_counts, color=['cyan', 'blue', 'darkblue'])
ax[0].set_title("Training Data Distribution")
ax[0].set_ylabel("Count")

ax[1].bar(test_labels, test_counts, color=['cyan', 'blue', 'darkblue'])
ax[1].set_title("Test Data Distribution")
ax[1].set_ylabel("Count")

plt.tight_layout()
plt.show()

# Classification Performance Data
models = ["BERT_No_Exp.", "BERT_Exp.", "XGBoost"]
accuracy = [0.59, 0.58, 0.30]
avg_f1_score = [0.57, 0.55, 0.28]

# Plot Model Comparison
fig, ax = plt.subplots(figsize=(8, 5))
x = np.arange(len(models))
width = 0.3

ax.bar(x - width/2, accuracy, width, label='Accuracy', color='blue')
ax.bar(x + width/2, avg_f1_score, width,
       label='Avg F1-score', color='darkblue')

ax.set_xticks(x)
ax.set_xticklabels(models)
ax.set_ylabel("Score")
ax.set_title("Model Performance Comparison")
ax.legend()

# Show values on top of bars
for bars in ax.containers:
    ax.bar_label(bars, fmt='%.2f', padding=3)

plt.tight_layout()
plt.show()
```
